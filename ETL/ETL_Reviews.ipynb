```python
import pandas as pd
import ast
```

```python
datos_reviews = []

with open("../Datasets_Originales/australian_user_reviews.json",encoding="utf-8") as reviews:
    for linea in reviews.readlines():
        datos_reviews.append(ast.literal_eval(linea))
        
for i in range(5):
    print(datos_reviews[i])
```

```python
df_reviews = pd.DataFrame(datos_reviews)
df_reviews
```

```python
df_reviews["user_id"].value_counts()
```

```python
df_reviews[df_reviews["user_id"] == "76561198027488037"]
```

```python
df_reviews = df_reviews.drop_duplicates(subset="user_id",keep="first")
df_reviews[df_reviews["user_id"] == "76561198027488037"]
```

```python
df_reviews["user_id"].value_counts()
```

```python
df_reviews_expandido= df_reviews.explode("reviews")
df_reviews_expandido
```

```python
df_reviews_desanidado = pd.json_normalize(df_reviews_expandido["reviews"])
df_reviews_desanidado
```

```python
df_reviews_desanidado.reset_index(drop=True,inplace=True)
df_reviews_expandido.reset_index(drop=True,inplace=True)
```

```python
df_reviews_final = pd.concat([df_reviews_expandido,df_reviews_desanidado],axis=1)
df_reviews_final
```

```python
a_eliminar = ["user_url","reviews","posted","last_edited","helpful","funny"]
df_reviews_final = df_reviews_final.drop(a_eliminar,axis=1)
```
