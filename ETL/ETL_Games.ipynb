### Etapa de Transformación

#### Analizando la informacion, sabemos que la columna app_name y title son los mismos datos,tambien hay una gran cantidad de columnas con datos nulos a lo que borraremos

```python
#Hacemos la limpieza de datos nulos
Juegos.dropna(how="all",inplace=True)
```

##### Vamos a comprobar los nulos de la columna app_name, y si encuentra nulos va a guardar el index de donde estan

```python
FilasaBorrar = Juegos[Juegos["app_name"].isnull()].index
Juegos[Juegos["app_name"].isnull()]
```

##### Ahora vamos a borrar las siguientes dos filas que tenemos sin mucha info

```python
Juegos.drop(index=FilasaBorrar,inplace=True)

Juegos[Juegos["app_name"].isnull()]
```

```python
comparacion = Juegos["app_name"].compare(Juegos["title"])
comparacion	
```

```python
comparacion[comparacion["other"].notna()]
```

##### notamos que en Other(title) no solo está con nulos sinó que tienen muchos errores en los nombres de cada titulo de los juegos. ahora vamos a borrar la columna title y tambien las columnas que no vamos a utilizar o no vamos a considerar utiles

```python
ColumnasAEliminar = ["title","url","tags","early_access","publisher","specs","reviews_url"]
Juegos.drop(columns=ColumnasAEliminar,inplace=True,axis=1)
Juegos
```

```python
##### hacemos un reseteo de index
# hacemos un reseteo del index
Juegos.reset_index(drop=True,inplace=True)
```

```python
index_id_vacio = Juegos[Juegos["id"].isna()].index
Juegos[Juegos["id"].isna()]
```

```python
Juegos.loc[index_id_vacio,["id"]] = 200260
Juegos.loc[index_id_vacio]
```

```python
Juegos.head(10)
```

```python
Juegos["id"] = Juegos["id"].astype(int)
Juegos
```

```python

def Normalizacion_Texto(valor):
    if isinstance(valor,str) and 'free'in valor.lower():
        return "free"
    elif isinstance(valor,str):
        return valor.lower()
    elif isinstance(valor,float):
        return str(round(valor,2))  
    else:
         return valor
       

Juegos["price"] = Juegos["price"].apply(Normalizacion_Texto)
```

```python
Juegos["price"].unique()
```

```python
Juegos.rename(columns={"id":"id_item"},inplace=True)
```

```python
Juegos["release_date"]= pd.to_datetime(Juegos["release_date"],errors="coerce")
```
